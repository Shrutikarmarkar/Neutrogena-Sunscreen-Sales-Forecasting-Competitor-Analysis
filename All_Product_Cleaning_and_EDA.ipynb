#Text Cleaning Pipeline for 'text' Column
##Steps:

1. Lowercase the text
2. Replace domain-specific phrases like “face wash” → “face-wash”
3. Remove symbols/punctuation (but preserve useful ones like hyphens)
4. Correct spelling (preserve domain terms)
5. Remove stopwords (but keep important ones like “not”, “very”, “the”)
6. Lemmatize → Stem
7. Remove rare words (occur only once across all reviews)
8. Join cleaned tokens into a string for later use

```python
!pip install pyspellchecker
```

```python
from spellchecker import SpellChecker
from collections import Counter
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
spell = SpellChecker()
domain_terms = {
    "face wash": "face-wash",
    "body wash": "body-wash",
    "sun screen": "sunscreen",
    "broad spectrum": "broad-spectrum",
    "moisturizing lotion": "moisturizing-lotion",
    "dry skin": "dry-skin",
    "oily skin": "oily-skin",
    "sensitive skin": "sensitive-skin",
    "anti aging": "anti-aging",
    "white cast": "white-cast",
    "quick absorb": "quick-absorb"
}
def clean_text(text):
    if not isinstance(text, str):
        return ""
    text = text.lower()
    for phrase, token in domain_terms.items():
        text = text.replace(phrase, token)
    text = re.sub(r"http\S+|<.*?>|\d+", "", text)
    text = re.sub(f"[{re.escape(string.punctuation)}]", "", text)
    tokens = nltk.word_tokenize(text)
    corrected_tokens = []
    for word in tokens:
        if word in domain_terms.values():
            corrected_tokens.append(word)
        else:
            try:
                corrected = spell.correction(word)
                corrected_tokens.append(corrected if corrected else word)
            except:
                corrected_tokens.append(word)
    #tokens = word_tokenize(text)
    filtered = [w for w in corrected_tokens if w not in stop_words and isinstance(w, str)]
    cleaned = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return " ".join(cleaned)

#df['ReviewText'] = df['text'].apply(clean_text)
#df.drop(columns=['text'], inplace=True)
```

```python
# Run LDA on Each Brand
from sklearn.decomposition import LatentDirichletAllocation

lda_Products = LatentDirichletAllocation(n_components=3, random_state=42)
lda_Products.fit(Products_dtm)

```

```python
# Display Topics
# Define a function to print topics
def print_topics(model, feature_names, n_top_words=10):
    for idx, topic in enumerate(model.components_):
        print(f"Topic {idx+1}:")
        print(" ".join([feature_names[i] for i in topic.argsort()[:-n_top_words - 1:-1]]))
        print()
print("🔹 Products - Negative Review Topics")
print_topics(lda_Products, Products_feature_names)

```

```python
# Daily aggregation of reviews
daily_reviews = df_reviews.groupby(df_reviews['ReviewDate'].dt.date).agg({
    'text': 'count',
    'Sentiment': lambda x: (x.map({'positive': 1, 'neutral': 0, 'negative': -1}).mean())
}).reset_index()

# Rename columns
daily_reviews.columns = ['Date', 'ReviewCount', 'AvgSentimentScore']

# Convert Date
daily_reviews['Date'] = pd.to_datetime(daily_reviews['Date'])

# Preview
daily_reviews.head()
```

```python
# Merge sales + reviews on Date
daily_combined = pd.merge(df_NeutrogenaSales, daily_reviews, on='Date', how='left')

# Fill missing values for ReviewCount and AvgSentimentScore
daily_combined['ReviewCount'] = daily_combined['ReviewCount'].fillna(0)
daily_combined['AvgSentimentScore'] = daily_combined['AvgSentimentScore'].fillna(0)

# Preview
daily_combined.head()
```

```python
# If Date is already the index, just confirm it
if 'Date' not in daily_combined.columns:
    daily_combined.index = pd.to_datetime(daily_combined.index)
else:
    daily_combined['Date'] = pd.to_datetime(daily_combined['Date'])
    daily_combined.set_index('Date', inplace=True)
```

```python
# First, set Date as index (already you did this correctly)
if 'Date' not in daily_combined.columns:
    daily_combined.index = pd.to_datetime(daily_combined.index)
else:
    daily_combined['Date'] = pd.to_datetime(daily_combined['Date'])
    daily_combined.set_index('Date', inplace=True)

# Group by Product, then Resample Weekly
weekly_list = []

for product, group in daily_combined.groupby('Product'):
    group = group.resample('W').agg({
        'Price': 'mean',
        'Best Seller Rank': 'mean',
        'ReviewCount': 'sum',
        'AvgSentimentScore': 'mean'
    })
    group['Product'] = product  # Add back Product column
    weekly_list.append(group)

# Combine all
weekly_combined = pd.concat(weekly_list).reset_index()

# Preview
weekly_combined.head()
```

```python
# Avoid 0 or missing prices
weekly_combined['Price'] = weekly_combined['Price'].replace(0, np.nan)  # First mark 0 as missing
weekly_combined['Price'] = weekly_combined['Price'].fillna(weekly_combined['Price'].mean())  # Then fill missing

# Now calculate SalesEstimate
weekly_combined['SalesEstimate'] = (weekly_combined['ReviewCount'] * weekly_combined['AvgSentimentScore']) / weekly_combined['Price']
# Make SalesEstimate always positive (optional)
#weekly_combined['SalesEstimate'] = weekly_combined['SalesEstimate'].clip(lower=0)
```

```python
# First, normalize SalesEstimate for comparison (already you did)
weekly_combined['SalesEstimateNorm'] = weekly_combined['SalesEstimate'] / weekly_combined['SalesEstimate'].max()

# Now plot separately for each product
plt.figure(figsize=(14,6))

for product in weekly_combined['Product'].unique():
    product_data = weekly_combined[weekly_combined['Product'] == product]
    plt.plot(product_data['Date'], product_data['SalesEstimate'], label=f'Product {product}')

plt.title('Estimated Weekly Sales Trend (Normalized) - Products A, B, C')
plt.xlabel('Week')
plt.ylabel('Normalized Sales Estimate')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
weekly_combined['EstimatedUnits'] = weekly_combined['SalesEstimate'] * 25
```

```python
plt.figure(figsize=(14,6))

# Plot separately for each product
for product in weekly_combined['Product'].unique():
    product_data = weekly_combined[weekly_combined['Product'] == product]
    plt.plot(product_data['Date'], product_data['EstimatedUnits'],  label=f'Product {product}')

plt.title('Estimated Weekly Units Sold - Products A, B, C')
plt.xlabel('Week')
plt.ylabel('Estimated Units')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# First, calculate 4-week moving average separately for each product
weekly_combined['TrendLine'] = weekly_combined.groupby('Product')['EstimatedUnits'].transform(lambda x: x.rolling(window=4, min_periods=1).mean())

# Now plot
plt.figure(figsize=(14,6))

for product in weekly_combined['Product'].unique():
    product_data = weekly_combined[weekly_combined['Product'] == product]
    plt.plot(product_data['Date'], product_data['TrendLine'], label=f'Product {product} - 4W Avg')

plt.title('4-Week Moving Average of Estimated Units Sold - Products A, B, C')
plt.xlabel('Week')
plt.ylabel('Estimated Units (4-Week Avg)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
